# **Importing Libraries**

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
sns.set(style="whitegrid", palette="muted")
```

# Explanation:

pandas and numpy: For data manipulation and numerical operations.

matplotlib and seaborn: For creating visualizations.

scipy.stats.pearsonr: For calculating Pearson's correlation coefficient.

Set the Seaborn style to create clean, attractive plots.


# Loading the GDP **Data**

 Loading the GDP dataset and displaying the first few rows.

```python
# Load the GDP dataset
gdp_data = pd.read_csv('/content/India_GDP_1960-2022.csv')

# Display the first few rows of the dataset
gdp_data.head()
```

# **Exploring the Data Structure**

 info() provides a summary of the dataset, including the column data types and any missing values.

```python
# Checking the structure and data types of the dataset
gdp_data.info()
```

# **Handling Missing Values**


```python
# Checking for missing values
gdp_data.isnull().sum()

# Handling missing values (dropping rows with missing GDP values)
gdp_data.dropna(subset=['GDP in (Billion) $'], inplace=True)
```

This code checks for missing values and drops rows where GDP data is missing to ensure clean data.

# **Ensuring Data Types**

```python
# Ensure that the 'Year' column is treated as an integer
gdp_data['year'] = gdp_data['year'].astype(int)
```

This ensures that the 'Year' column is in the correct integer format for time-series analysis.


# **Visualizing GDP Trend (1960-2022)**

```python
# Plotting India's GDP over time (1960-2022)
plt.figure(figsize=(10, 6))
plt.plot(gdp_data['year'], gdp_data['GDP in (Billion) $'], marker='o', color='blue')
plt.title("India's GDP from 1960 to 2022")
plt.xlabel("Year")
plt.ylabel("GDP (in billion USD)")
plt.grid(True)
plt.show()
```

 Visualizes the trend of India's GDP over time using a line plot.


# **Descriptive Statistics of GDP**

```python
# Getting summary statistics for GDP
gdp_data['GDP in (Billion) $'].describe()

```

This code provides summary statistics (mean, median, standard deviation, etc.) for the GDP column.


# **Calculating Year-over-Year GDP Growth Rate**

```python
# Calculating the year-over-year GDP growth rate
gdp_data['GDP_Growth_Rate'] = gdp_data['GDP in (Billion) $'].pct_change() * 100

# Dropping the first row (NaN growth rate)
gdp_data.dropna(subset=['GDP_Growth_Rate'], inplace=True)
```


The pct_change() function calculates the percentage change in GDP from one year to the next, giving us the year-over-year growth rate.


# **Visualizing Year-over-Year GDP Growth Rate**

```python
# Plotting the year-over-year GDP Growth Rate
plt.figure(figsize=(10, 6))
plt.plot(gdp_data['year'], gdp_data['GDP_Growth_Rate'], marker='o', color='green')
plt.title("India's Year-over-Year GDP Growth Rate (1961-2022)")
plt.xlabel("Year")
plt.ylabel("GDP Growth Rate (%)")
plt.grid(True)
plt.show()
```

Visualizes the year-over-year GDP growth rate for India, highlighting periods of rapid economic growth or contraction.

# **Checking the Distribution of GDP Growth Rates**

```python
# Plotting the distribution of GDP growth rates
plt.figure(figsize=(8, 5))
sns.histplot(gdp_data['GDP_Growth_Rate'], kde=True, color='green')
plt.title("Distribution of GDP Growth Rates (1961-2022)")
plt.xlabel("GDP Growth Rate (%)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

This histogram, with a KDE plot overlaid, shows the distribution of GDP growth rates, helping to identify how growth has fluctuated.

# **Calculating Correlation Between GDP and Growth Rate**

```python
# Calculating Pearson correlation between GDP and GDP Growth Rate
corr_gdp_growth, p_value_gdp_growth = pearsonr(gdp_data['GDP in (Billion) $'], gdp_data['GDP_Growth_Rate'])

# Display the correlation results
print(f"Pearson Correlation Coefficient: {corr_gdp_growth:.2f}")
print(f"P-value: {p_value_gdp_growth:.4f}")
```

Pearson's correlation coefficient is calculated to quantify the linear relationship between GDP levels and GDP growth rate.


# **Interpreting the Correlation Results**

Pearson Correlation Coefficient: corr_gdp_growth:.2f, indicating the strength and direction of the relationship between GDP and its growth rate.

P-value: p_value_gdp_growth:.4f, indicating whether the correlation is statistically significant (typically if p-value < 0.05).


# **Scatter Plot of GDP vs. Growth Rate**

```python
# Scatter plot of GDP vs. GDP Growth Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GDP in (Billion) $', y='GDP_Growth_Rate', data=gdp_data, color='purple')
plt.title("Relationship Between GDP and GDP Growth Rate")
plt.xlabel("GDP (in billion USD)")
plt.ylabel("GDP Growth Rate (%)")
plt.grid(True)
plt.show()
```

 A scatter plot visually represents the relationship between GDP and its growth rate, allowing us to observe any visible correlation.

# **Additional Code and Insights:**

```python
# Calculating a rolling average of the GDP growth rate (window of 5 years)
gdp_data['Rolling_Avg_Growth'] = gdp_data['GDP_Growth_Rate'].rolling(window=5).mean()

# Visualizing the rolling average of GDP growth rate
plt.figure(figsize=(10, 6))
plt.plot(gdp_data['year'], gdp_data['Rolling_Avg_Growth'], marker='o', color='orange')
plt.title("5-Year Rolling Average of GDP Growth Rate (1965-2022)")
plt.xlabel("Year")
plt.ylabel("Rolling Avg. Growth Rate (%)")
plt.grid(True)
plt.show()
```

Rolling Average of GDP Growth: Helps smooth out short-term fluctuations.
Auto-Correlation: Check if current GDP depends on past values (lagged correlation).
